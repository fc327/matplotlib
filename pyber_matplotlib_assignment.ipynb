```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load 
city_data_path = ("../Pyber/data/city_data.csv")
ride_data_path = ("../Pyber/data/ride_data.csv")
```

```python
# Read the City and Ride Data
city_df=pd.read_csv(city_data_path)
ride_df=pd.read_csv(ride_data_path)
```

```python
# Combine the data into a single dataset
rides_and_cities_df=pd.merge(city_df, ride_df, how="outer", on="city", left_on = None, right_on = None)
```

```python
# Display the data table for preview
rides_and_cities_df.head()
```

```python
urban = rides_and_cities_df[rides_and_cities_df['type']=='Urban']
urban
suburban = rides_and_cities_df[rides_and_cities_df['type']=='Suburban']
suburban
rural = rides_and_cities_df[rides_and_cities_df['type']=='Rural']
rural
```

```python
urban_count= urban.groupby('city')['ride_id'].count()
urban_count
suburban_count= suburban.groupby('city')['ride_id'].count()
suburban_count
rural_count= rural.groupby('city')['ride_id'].count()
rural_count
```

```python
# Obtain the x and y coordinates for each of the three city types


# Build the scatter plots for each city types
plt.scatter(x=urban_count, y=urban_avg_fare, s=urban_driver_count*10, alpha=0.8, c='coral', edgecolor='black', linewidths=1, marker='o')
plt.scatter(x=suburban_count, y=suburban_avg_fare, s=suburban_driver_count*10, alpha=0.8, c='skyblue', edgecolor='black', linewidths=1, marker='o')
plt.scatter(x=rural_count, y=rural_avg_fare, s=rural_driver_count*10, alpha=0.8, c='gold', edgecolor='black', linewidths=1, marker='o')
# Incorporate the other graph properties


# Create a legend
city_legend= plt.legend(('urban', 'suburban', 'rural'), ncol = 1, loc = 'upper right', fontsize = 10, title='city type')

city_legend.legendHandles[0]._sizes = [30]
city_legend.legendHandles[1]._sizes = [30]
city_legend.legendHandles[2]._sizes = [30]

plt.xlabel('Total Number of Rides (per city)')
plt.ylabel('Average Fare($)')
plt.title('Pyber Ride Sharing Data from 2016')

plt.grid(True)

# Incorporate a text label regarding circle size
plt.text(45,35, 'Note: circle sizes correlates with driver count per city')

# Save Figure
plt.show()
```
